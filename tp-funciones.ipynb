Una pista, la función ord() convierte un carácter en un su código ASCII y la función chr() convierte el código ASCII en carácter.
```python
#Nicolas Pecora.

#Ejercicio 3

#Defino Funciones

def verificacion_str(opcion,a,b):
    """Solicita una entrada y verifica si son iguales a las dos constantes ingresadas.
    si no coincide solicita una entrada nueva hasta que se cumpla la condicion.
    Entrada: entrada usuario(str), comparacion(str), comparacion(str)
    Salida: la opcion(str) coincidente con la condicion"""
    verificacion = False
    while verificacion == False:
        if opcion.upper() == a or opcion.upper() == b:
            verificacion = True
        else:
            opcion = input("ATENCION: Entrada incorrecta, reingrese opcion: ")
    return opcion

def lee_entero(clave):
    """ Solicita un valor entero y lo devuelve.
        Mientras el valor ingresado no sea entero, vuelve a solicitarlo. """
    ver_num = False
    while ver_num == False:
        try:
            clave = int(clave)
            ver_num = True
            return clave
        except ValueError:
            clave = input("ATENCIÓN: Debe ingresar un número entero: ")
    
    
def encriptado(frase, clave,):
    """
    Función encriptado: Encripta un frase 
    cambiando el valor de los caracteres segun su clave.
    Devuelve un una frase encriptada.
    Entrada: frase(string), clave(int)
    Salida: (frase_encriptada(string))
    """
    _123 = "0123456789"
    abc = "abcdefghijklmnñopqrstuvwxyz"
    ABC = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    frase_encriptada = ""
    for letra in frase:
        if letra in abc:
            lugar_letra = abc.find(letra)
            lugar_letra += clave
            if lugar_letra >= 27:
                while lugar_letra >= 27:
                    lugar_letra -= 26
                frase_encriptada += abc[lugar_letra - 1]
            else:
                frase_encriptada += abc[lugar_letra]
        elif letra in ABC:
            lugar_letra = ABC.find(letra)
            lugar_letra += clave
            if lugar_letra >= 27:
                while lugar_letra >= 27:
                    lugar_letra -= 26
                frase_encriptada += ABC[lugar_letra - 1]
            else:
                frase_encriptada += ABC[lugar_letra]
        elif letra in _123:
            lugar_letra = _123.find(letra)
            lugar_letra += clave
            if lugar_letra >= 11:
                while lugar_letra >= 11:
                    lugar_letra -= 10
                frase_encriptada += _123[lugar_letra - 1]  
            else:
                frase_encriptada += _123[lugar_letra]
        else:
            frase_encriptada += letra
    return frase_encriptada

def desencriptado(frase, clave,):
    """
    Función desencriptado: Desencripta una 
    frase cambiando el valor de los caracteres segun su clave.
    Devuelve un una frase desencriptada.
    Entrada: frase(string), clave(int)
    Salida: (frase_encriptada(string))
    """
    _123 = "0123456789"
    abc = "abcdefghijklmnñopqrstuvwxyz"
    ABC = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

    frase_desencriptada = ""

    for letra in frase:
        if letra in abc:
            lugar_letra = abc.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -27:
                    lugar_letra += 26
                frase_desencriptada += abc[lugar_letra + 27]
            else:
                frase_desencriptada += abc[lugar_letra + 26]
        elif letra in ABC:
            lugar_letra = ABC.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -28:
                    lugar_letra += 26
                frase_desencriptada += ABC[lugar_letra + 27]
            else:                    
                frase_desencriptada += ABC[lugar_letra + 26]
        elif letra in _123:
            lugar_letra = _123.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -11:
                    lugar_letra += 10
                frase_desencriptada += _123[lugar_letra + 11]
            else:
                frase_desencriptada += _123[lugar_letra + 10]
        else:
            frase_desencriptada += letra
    return frase_desencriptada 

def en_des(frase, clave,):
    """
    Función en_des: Encripta o desencripta una frase cambiando 
    el valor de los caracteres segun su clave.
    Devuelve una frase encriptada y desencriptada en una lista.
    Entrada: frase(string), clave(int)
    Salida: [frase encriptada(string), frase desencriptada(string)]
    """
    _123 = "0123456789"
    abc = "abcdefghijklmnñopqrstuvwxyz"
    ABC = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    frase_encriptada = ""
    for letra in frase:
        if letra in abc:
            lugar_letra = abc.find(letra)
            lugar_letra += clave
            if lugar_letra >= 27:
                while lugar_letra >= 27:
                    lugar_letra -= 26
                frase_encriptada += abc[lugar_letra - 1]
            else:
                frase_encriptada += abc[lugar_letra]
        elif letra in ABC:
            lugar_letra = ABC.find(letra)
            lugar_letra += clave
            if lugar_letra >= 27:
                while lugar_letra >= 27:
                    lugar_letra -= 26
                frase_encriptada += ABC[lugar_letra - 1]
            else:
                frase_encriptada += ABC[lugar_letra]
        elif letra in _123:
            lugar_letra = _123.find(letra)
            lugar_letra += clave
            if lugar_letra >= 11:
                while lugar_letra >= 11:
                    lugar_letra -= 10
                frase_encriptada += _123[lugar_letra - 1]  
            else:
                frase_encriptada += _123[lugar_letra]
        else:
            frase_encriptada += letra

    frase_desencriptada = ""
    for letra in frase:
        if letra in abc:
            lugar_letra = abc.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -27:
                    lugar_letra += 26
                frase_desencriptada += abc[lugar_letra + 27]
            else:
                frase_desencriptada += abc[lugar_letra + 26]
        elif letra in ABC:
            lugar_letra = ABC.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -28:
                    lugar_letra += 26
                frase_desencriptada += ABC[lugar_letra + 27]
            else:                    
                frase_desencriptada += ABC[lugar_letra + 26]
        elif letra in _123:
            lugar_letra = _123.find(letra)
            lugar_letra -= clave
            if lugar_letra < 0:
                while lugar_letra < -11:
                    lugar_letra += 10
                frase_desencriptada += _123[lugar_letra + 11]
            else:
                frase_desencriptada += _123[lugar_letra + 10]
        else:
            frase_desencriptada += letra
    
    return [frase_encriptada, frase_desencriptada]

#Programa principal
print("""*********************************************************
ENCRIPTADOR Y DESENCRIPTADOR DE FRASES (ENCRIPTADO CESAR)
*********************************************************"""
    )

# Loop para continuar con el programa
seguir = "S"
while seguir.upper() == "S":

# Ingreso de la frase
    frase = input("Ingrese una frase: ")

# Ingreso de la clave

    clave = lee_entero(input("ingrese clave: "))
    

# Ingreso encriptado
    
    pregunta = verificacion_str(input("ingrese opcion: (E)Encriptar, (D)Desencriptar: "),"E","D")

# Ingreso Tipo de funcion a utilizar
    
    metodo = verificacion_str(input("ingrese metodo (1)Solo una funcion, (2)Dos funciones: "),"1","2")      

# opcion del metodo a utilizar
    if metodo == "1":
        if pregunta.upper() == "E":
            print(f"Elegiste metodo 1:La frase encriptada es: {en_des(frase,clave)[0]}")
        else:
            print(f"Elegiste metodo 1:La frase desencriptada es: {en_des(frase,clave)[1]}")
    else:
        if pregunta.upper() == "E":
            print(f"Elegiste metodo 2:La frase encriptada es: {encriptado(frase,clave)}")
        else:
            print(f"Elegiste metodo 2:La frase desencriptada es: {desencriptado(frase,clave)}")

# Ingreso continuar el programa
   
    seguir = verificacion_str(input("¿desea hacer otra comprobacion? (s)SI//(n)NO: "),"S","N")
        
print("Gracias por usar el Encriptador y Desencriptador de frases")




```
